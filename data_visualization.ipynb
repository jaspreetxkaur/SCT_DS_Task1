```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## Dataset Creation
##Here we create a synthetic dataset for visualization purposes.
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Daisy', 'Eva', 'Frank', 'Grace', 'Henry', 'Isla', 'Jack'],
    'Age': [23, 25, 30, 22, 28, 35, 26, 29, 31, 27],
    'Score': [85, 78, 90, 88, 76, 95, 89, 84, 91, 80],
    'Gender': ['F', 'M', 'M', 'F', 'F', 'M', 'F', 'M', 'F', 'M'],
    'Region': ['North', 'South', 'East', 'West', 'North', 'South', 'East', 'West', 'North', 'South']
}

df = pd.DataFrame(data)
df

```

```python
## Basic Visualizations
### 1. Bar Chart - Scores by Name
plt.figure(figsize=(8,5))
sns.barplot(x='Name', y='Score', data=df, palette='viridis')
plt.title('Scores by Name')
plt.xticks(rotation=45)
plt.show()
```

```python
### 2. Histogram - Age Distribution
```

```python
plt.figure(figsize=(7,4))
plt.hist(df['Age'], bins=5, color='orange', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
### 3. Scatter Plot - Age vs Score
```

```python
plt.figure(figsize=(7,4))
sns.scatterplot(x='Age', y='Score', hue='Gender', data=df, s=100)
plt.title('Age vs Score by Gender')
plt.show()
```
